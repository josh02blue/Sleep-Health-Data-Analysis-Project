```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
sleep_data = pd.read_csv("Sleep_health_and_lifestyle_dataset.csv")
sleep_data
```

## Occupation Column

```python
occupation_count = sleep_data["Occupation"].value_counts()
occupation_count.plot(kind = "bar")
plt.title("Distribution of Occupation")
plt.xlabel("Occupation")
plt.ylabel("# of people")
```

## BMI Category Column

```python
bmi_category_count = sleep_data["BMI Category"].value_counts()
bmi_category_count.plot(kind = "bar")
plt.title("Distribution of BMI")
plt.xlabel("BMI Categories")
plt.ylabel("# of people")
```

## Age Column

```python
sleep_data["Age"].hist()
plt.title("Distribution of Age")
plt.xlabel("Age")
plt.ylabel("# of people")
```

## Stress Level Column

```python
sleep_data["Stress Level"].hist(bins = 15)
plt.title("Distribution of Stress Level")
plt.xlabel("Stress Levels")
plt.ylabel("# of people")
```
