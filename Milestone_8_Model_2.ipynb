```python
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.formula.api as smf
import seaborn as sns
import numpy as np

from sklearn.model_selection import train_test_split

from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier

from sklearn.metrics import confusion_matrix
```

```python
sleep_data = pd.read_csv("./Sleep_health_and_lifestyle_dataset.csv")
sleep_data.describe(include = "all")
```

```python
sleep_data_2 = sleep_data[['Age', 'Heart Rate', 'Sleep Disorder']].copy()
```

```python
sleep_data_2['Sleep Disorder'] = sleep_data_2['Sleep Disorder'].fillna('None')
```

```python
sleep_data_2['Sleep Disorder'] = sleep_data_2['Sleep Disorder'].map({
    'None': 0,
    'Sleep Apnea': 1,
    'Insomnia': 2
})
```

```python
sleep_data_2
```

```python
x = sleep_data_2[['Age', 'Heart Rate']]
y = sleep_data_2['Sleep Disorder']
```

### b) Prepare your data for machine learning: split your data into training and testing data, and if necessary, create dummy variables or scale columns.

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

### d) Evaluate the fit of your model using your testing and training data. You should use the appropriate methods for your model.

### Random Forest Classifier

```python
rf_50 = RandomForestClassifier(n_estimators = 10, max_depth = 2)
rf_50.fit(x_train, y_train)
```

```python
y_test_pred_rf_50 = rf_50.predict(x_test)
```

```python
confusion_matrix(y_test,y_test_pred_rf_50)
```

```python
from sklearn.metrics import accuracy_score

rf_50_accuracy = accuracy_score(y_test, y_test_pred_rf_50)
print("Accuracy:", rf_50_accuracy)
```
