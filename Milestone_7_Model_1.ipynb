```python
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.formula.api as smf
import seaborn as sns
import numpy as np
```

### a) Choose a column in your dataset to predict using some or all of the other columns.. This can either be a quantitative column to perform regression on, or a qualitative column to perform classification on.

```python
sleep_data = pd.read_csv("Sleep_health_and_lifestyle_dataset.csv")
sleep_data.describe(include = "all")
```

### b) Prepare your data for machine learning: split your data into training and testing data, and if necessary, create dummy variables or scale columns.

```python
sleep_data_2 = sleep_data[['Age', 'Heart Rate', 'Sleep Disorder']].copy()
```

```python
sleep_data_2['Sleep Disorder'] = sleep_data_2['Sleep Disorder'].fillna('None')
```

```python
sleep_data_2['Sleep Disorder'] = sleep_data_2['Sleep Disorder'].map({
    'None': 0,
    'Sleep Apnea': 1,
    'Insomnia': 2
})
```

```python
sleep_data_2
```

```python
x = sleep_data_2[['Age', 'Heart Rate']]
y = sleep_data_2['Sleep Disorder']
```

### b) Prepare your data for machine learning: split your data into training and testing data, and if necessary, create dummy variables or scale columns.

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split


x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

### d) Evaluate the fit of your model using your testing and training data. You should use the appropriate methods for your model.

```python
tree_3 = DecisionTreeClassifier(max_depth = 3)
tree_3.fit(x_train, y_train)
```

```python
y_test_pred_3 = tree_3.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, y_test_pred_3)
```

```python
from sklearn.metrics import accuracy_score

accuracy_3 = accuracy_score(y_test, y_test_pred_3)
print("Accuracy:", accuracy_3)
```
