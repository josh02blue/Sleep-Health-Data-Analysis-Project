```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
sleep_data = pd.read_csv("Sleep_health_and_lifestyle_dataset.csv")
sleep_data
```

## Occupation vs Stress Level

```python
sns.boxplot(x = "Occupation", y = "Stress Level", data = sleep_data)
plt.title("Stress Levels by Occupation")
plt.xticks(rotation=90);
```

## Daily Steps vs Heart Rate

```python
sns.relplot(x = "Daily Steps", y = "Heart Rate",hue = "Gender", data = sleep_data)
plt.title("Daily Steps vs Heart Rate by Gender")
```
